```python
import numpy as np
import matplotlib.pyplot as plt
import sympy as sp 
```

```python
x=sp.symbols('x')
```

```python

def calculateGradientFunction(function):
    return sp.diff(function,x)

```

```python
def calculateActualGradient(gradient_function,value):
    return gradient_function.subs(x,value)
```

```python
def gradient_ascent(function,start_value=0.0,learning_rate=0.1,iterations=1000):
    gradient_function=calculateGradientFunction(function=function)
    theta=start_value
    gradients=[]
    thetas=[]
    function_vals=[]
    for i in range(iterations):
        # calculate the value of the gradient
        val=calculateActualGradient(gradient_function=gradient_function,value=theta)
        gradients.append(val)
        # applying the update rule
        theta=theta+learning_rate* (val)
        thetas.append(theta)

        # calculating the function value keeping the theta value 
        function_vals.append(function.subs(x,theta))


    return gradients,thetas,function_vals,theta
```

```python
# calculating basic differentials 
w=(-(x-3)**2)+4
start=0.0
alpha=0.1
iterations=1000

gradients,thetas,function_vals,final_val=gradient_ascent(function=w,start_value=start,learning_rate=alpha,iterations=iterations)
```

```python
for i in range(len(gradients)):
    print(f"The value of the gradient is {gradients[i]} and the value of theta is {thetas[i]} and the value of function is {function_vals[i]}")

print(f"The final converged theta is {final_val}")
```
